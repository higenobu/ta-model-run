```python
import re
r = (
    None,
    re.compile(r'^(.+) - 1.[\s\t]*喜びを感じた$'),
    re.compile(r'^(.+) - 2.[\s\t]*恐怖を感じた$'),
    re.compile(r'^(.+) - 3.[\s\t]*驚きを感じた$'),
    re.compile(r'^(.+) - 4.[\s\t]*信頼できる情報と感じた$'),
    re.compile(r'^(.+) - 5.[\s\t]*曖昧な情報と感じた$'),
    re.compile(r'^(.+) - 6.[\s\t]*何かの意図をもって書かれたと感じた$'),
    re.compile(r'^(.+) - 7.[\s\t]*経済に期待がもてると感じた$'),
)
repl = lambda m: m.group(1)
```

```python
df=dataframes[0]

def transform_dataframe(df):
    x=1200
    # 数値を抽出するための正規表現パターン
    pattern = r'(\d+)'

    # 数値部分を抽出して新しい列を作成
    # 整数型の値として解釈できるものを選択し、x 秒以上の行を削除
    df['Duration_numeric'] = pd.to_numeric(df['Duration (in seconds)'], errors='coerce')
    # x秒以上の行を削除
    df = df[(df['Duration_numeric'] > x)|(df['Duration_numeric'].isna())]
    data=df.T.iloc[4:,:]
    data = data[data.index!= 'Duration_numeric']
    data['ids'] = data.index.str.split(r'_', expand=False)

    # 新しい列にYの値を格納
    data['id_range'] = data['ids'].apply(lambda s: s[0])
    data['seq'] = data['ids'].apply(lambda s: s[1])
    data['text'] = data[0]
    for i in range(1, 8):
            data['text'] = data['text'].str.replace(r[i], repl, regex=True)
    return data
```

```python
import re
from pyknp import Juman
from tqdm import tqdm
import mojimoji

# 形態素解析器のインスタンスを作成
jumanpp = Juman()

# tqdmをPandasに組み込む
tqdm.pandas()

# テキストを文単位に分割し、【】で囲まれている部分を削除、全角変換、形態素解析を行う
def process_text(text):
    # テキストを全角に変換
    text = mojimoji.han_to_zen(text)
    
    # 【】で囲まれている部分を削除
    text = re.sub('【.*?】', '', text)

    # テキストを文単位に分割
    sentences = re.split('。|\n', text)

    # 文が20文字未満のものを除外
    sentences = [sentence for sentence in sentences if len(sentence) >= 20]

    # 各文に形態素解析を適用
    words = [' '.join([mrph.midasi for mrph in jumanpp.analysis(sentence.replace('^', '＾')).mrph_list()]) for sentence in sentences]

    return ' '.join(words)

combined_df['text'] = combined_df['text'].progress_apply(process_text)
```

```python
#seqの値によって、dfを分解
combined_df_seq=[[]for _ in range(8)]
for i in range(1,8):
    combined_df_seq[i]=combined_df[combined_df["seq"]==i]
    print(len(combined_df_seq[i]))
```

```python
from sklearn.model_selection import train_test_split
sentences_train_old_all=[[]for _ in range(8)]
labels_train_old_all=[[]for _ in range(8)]
sentences_dev_old_all=[[]for _ in range(8)]
labels_dev_old_all=[[]for _ in range(8)]
sentences_test_old_all=[[]for _ in range(8)]
labels_test_old_all=[[]for _ in range(8)]

for i in range(1,8):

    # 訓練データ、検証データ、テストデータに分割する
    train_val, test = train_test_split(combined_df_seq[i], test_size=0.1, random_state=42)
    train, val = train_test_split(train_val, test_size=0.111, random_state=42)

    # 訓練データ、検証データ、テストデータのサイズを表示する
    print(f"訓練データのサイズ: {len(train)}, 検証データのサイズ: {len(val)}, テストデータのサイズ: {len(test)}")
    
    for index, row in train.iterrows():
        sentences_train_old_all[i].append(row['text'])
        labels_train_old_all[i].append(row['mean'])
    for index, row in val.iterrows():
        sentences_dev_old_all[i].append(row['text'])
        labels_dev_old_all[i].append(row['mean'])
    for index, row in test.iterrows():
        sentences_test_old_all[i].append(row['text'])
        labels_test_old_all[i].append(row['mean'])
```

```python
import json

# Combine sentences and labels into a list of dictionaries
train_data = []
for i in range(1, 8):
	sentences = sentences_train_old_all[i]
	labels = labels_train_old_all[i]
	# make sure the labels are floats
	combined_data = [{'sentence': sentence, 'label': float(label)} for sentence, label in zip(sentences, labels)]
	train_data.append(combined_data)

# Save the data as JSON files
for i in range(1, 8):
	filename = f'train_data_{i}.json'
	with open(filename, 'w') as f:
		json.dump(train_data[i-1], f, ensure_ascii=False)

dev_data = []
for i in range(1, 8):
	sentences = sentences_dev_old_all[i]
	labels = labels_dev_old_all[i]
	combined_data = [{'sentence': sentence, 'label': float(label)} for sentence, label in zip(sentences, labels)]
	dev_data.append(combined_data)

# Save the data as JSON files
for i in range(1, 8):
	filename = f'dev_data_{i}.json'
	with open(filename, 'w') as f:
		json.dump(dev_data[i-1], f, ensure_ascii=False)

test_data = []
for i in range(1, 8):
	sentences = sentences_test_old_all[i]
	labels = labels_test_old_all[i]
	combined_data = [{'sentence': sentence, 'label': float(label)} for sentence, label in zip(sentences, labels)]
	test_data.append(combined_data)

# Save the data as JSON files
for i in range(1, 8):
	filename = f'test_data_{i}.json'
	with open(filename, 'w') as f:
		json.dump(test_data[i-1], f, ensure_ascii=False)
```
